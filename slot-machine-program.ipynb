```python
import random
def spin_row():
    symbols = ['ðŸ’','ðŸ‰','ðŸ””','ðŸ‹']

    return [random.choice(symbols) for _ in range(3)]

def print_row(row):
    print(" | ".join(row))

def get_payout(row, bet):
    if row[0] == row[1] == row[2]:
        if row[0] == 'ðŸ’':
            return bet * 3
        elif row[0] == 'ðŸ‰':
            return bet * 4
        elif row[0] == 'ðŸ””':
            return bet * 5
        elif row[0] == 'ðŸ‹':
            return bet * 6
    return 0

def main():
    balance = 100

    print("Welcome to Python Slots")
    print("Symbols: ðŸ’ðŸ‰ðŸ””ðŸ‹")

    while balance > 0:
        print(f"Current balance: ${balance}")

        bet = float(input("Place your bet amount"))

        bet = int(bet)

        if bet > balance:
            print("Insufficient funds")
            continue

        if bet <= 0:
            print("Bet must be greater than 0")
            continue

        balance -= bet

        row = spin_row()
        print("Spinning...\n")
        print_row(row)

        payout = get_payout(row, bet)

        if payout > 0:
            print(f"You won ${payout}")
        else:
            print("Sorry you lost this round")

        balance += payout

        play_again = input("Do you want to spin again? (Y/N): ").upper()

        if play_again != 'Y':
            break

if __name__ == '__main__':
    main()
```
